# seaborn: statistical data visualization
Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.

you can ref [seaborn galery](http://seaborn.pydata.org/examples/) and [seaborn tutorial](http://seaborn.pydata.org/tutorial.html) for more

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv('datasets/Pokemon.csv', index_col=0,encoding="utf8")
df.head()
```

## plotting functions

```python
sns.lmplot(x='Attack', y='Defense', data=df)
```

```python
sns.lmplot(x='Attack', y='Defense', data=df,
           fit_reg=False, # No regression line
           hue='Stage')  # Color by evolution 'stage' in df
```

## Customizing with Matplotlib

```python
sns.lmplot(x='Attack', y='Defense', data=df,
           fit_reg=False, 
           hue='Stage')

# Tweak using Matplotlib
plt.ylim(0, 200)
plt.xlim(0, 160)
```

## with Pandas

```python
sns.boxplot(data=df)
```

```python
stats_df = df.drop(['Total', 'Stage', 'Legendary'], axis = 1)

sns.boxplot(data = stats_df)
```

## Seaborn themes

```python
plt.style.available
```

```python
sns.set_style('whitegrid')

sns.violinplot(x='Type 1', y='Attack', data=df)
plt.rcParams['figure.figsize'] = [12, 8]
```

## Color palettes
Let's use [Bulbapedia](https://bulbapedia.bulbagarden.net/wiki/Category:Type_color_templates) to help us create a new color palette:

```python
pkmn_type_colors = ['#78C850',  # Grass
                    '#F08030',  # Fire
                    '#6890F0',  # Water
                    '#A8B820',  # Bug
                    '#A8A878',  # Normal
                    '#A040A0',  # Poison
                    '#F8D030',  # Electric
                    '#E0C068',  # Ground
                    '#EE99AC',  # Fairy
                    '#C03028',  # Fighting
                    '#F85888',  # Psychic
                    '#B8A038',  # Rock
                    '#705898',  # Ghost
                    '#98D8D8',  # Ice
                    '#7038F8',  # Dragon
                   ]
```

```python
sns.violinplot(x='Type 1', y='Attack', data=df, palette=pkmn_type_colors)
```

```python
sns.swarmplot(x='Type 1', y='Attack', data=df, palette=pkmn_type_colors)
```

## Overlaying plots

```python
plt.figure(figsize=(10,6))

sns.violinplot(x='Type 1', 
                y='Attack',
                data=df,
                inner=None,  # Remove the bars inside the violins
                palette=pkmn_type_colors)

sns.swarmplot(x='Type 1', 
                y='Attack',
                data=df,
                color='k', # Make points black
                alpha=0.7) # and slightly transparent

plt.title('Attack by Type')
```

## Putting it all together

```python
stats_df.head()
```

```python
melted_df = pd.melt(stats_df,
                   id_vars=["Name", "Type 1", "Type 2"], # Variables to keep
                   var_name="Stat") # Name of melted variable
melted_df.head()
```

```python
print(stats_df.shape)
print(melted_df.shape)

#please try to understand pandas function melt()
```

```python
sns.swarmplot(x='Stat', y='value', data=melted_df, hue="Type 1")
```

```python
plt.figure(figsize=(10,6))

sns.swarmplot(x='Stat',
              y='value', 
              data=melted_df, 
              hue="Type 1", 
              dodge=True, 
              palette=pkmn_type_colors)

plt.ylim(0, 260)

plt.legend(bbox_to_anchor=(1,1), loc=2)
```

## mini-gallery

### Heatmap

```python
corr = stats_df.corr()

sns.heatmap(corr)
```

### Histogram

```python
sns.distplot(df.Attack)
```

### Bar Plot

```python
sns.countplot(x="Type 1", data=df, palette=pkmn_type_colors)

plt.xticks(rotation=45)
```

```python
sns.catplot(x='Type 1', y='Attack', data=df, kind='bar',height=6, aspect=2, palette="muted", legend=False)
```

### Cat Plot
Cat plots make it easy to **separate plots** by categorical classes.

```python
g = sns.catplot(x='Type 1', 
                   y='Attack', 
                   data=df, 
                   hue='Stage',  # Color by stage
                   col='Stage',  # Separate by stage
                   kind='swarm') # Swarmplot

g.set_xticklabels(rotation=-45)
```

### Density Plot
Density plots display the distribution between two variables.

```python
sns.kdeplot(df.Attack, df.Defense)
```

### Joint Distribution Plot
Joint distribution plots combine information from scatter plots and histograms to give you detailed information for bi-variate distributions.

```python
sns.jointplot(x='Attack', y='Defense', data=df)
```
